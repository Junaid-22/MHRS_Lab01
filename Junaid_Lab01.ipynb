```python
import rasterio
from rasterio.plot import reshape_as_image
import cv2
try:
    img = rasterio.open("D:/Sir Ehtsham/Python/Classification.tif")
    img_array = img.read()
    img_reshaped = reshape_as_image(img_array)
    height, width = img_array.shape[1], img_array.shape[2]
    new_width = int(width / 2)
    new_height = int(height / 2)
    res = cv2.resize(img_reshaped, (new_width, new_height), interpolation=cv2.INTER_CUBIC)
    cv2.imwrite('result.tif', res)  
except IOError:
    print('Error while reading or processing files !!!')
```

```python
plt.imshow(res)
```

```python
try:
    img = rasterio.open("D:/Sir Ehtsham/Python/Classification.tif")
    img_array = img.read()
    img_reshaped = reshape_as_image(img_array)
    rows, column = img_array.shape[1], img_array.shape[2]
    new_rows = int(rows/ 2)
    new_column = int(column / 2)
    M = cv2.getRotationMatrix2D ((new_rows, new_column), 45,1)
    res1 = cv2.warpAffine(img_reshaped, M, (column, rows))
    cv2.imwrite('result.tif', res1)  
except IOError:
    print('Error while reading or processing files !!!')
```

```python
plt.imshow(res1)
```

```python
M1 = np.float32([[1, 0, 100], [0, 1, 50]])
try:
    res2 = cv2.warpAffine(img_reshaped, M1, (column, rows))
    cv2.imwrite('result.tiff', res2)
except IOError:
    print('Error while reading files !!!')
```

```python
plt.imshow(res2)
```

```python
try:
    edges = cv2.Canny(img_reshaped, 1, 2)
 
    # Write image back to disk.
    cv2.imwrite('result.tiff', edges)
 
except IOError:
    print('Error while reading files !!!')
```

```python
plt.imshow(edges)
```
